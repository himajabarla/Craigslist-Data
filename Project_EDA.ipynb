```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white")
```

```python
df = pd.read_sas('vehicles7_eda.sas7bdat')
```

```python
df.columns
```

```python
sns.pairplot(df)
```

```python
#Checking for normality of the dependent variable price
sns.set_palette("husl")
sns.set_style('whitegrid')
plt.figure(figsize=(12,7))
sns.distplot(df['price'])
plt.xlabel('price')
plt.ylabel('Prob density')
plt.title('Distribution plot of price',fontsize=15)
```

```python
#Checking for normality of age
sns.set_palette("husl")
sns.set_style('whitegrid')
plt.figure(figsize=(12,7))
sns.distplot(df['age'])
plt.xlabel('age')
plt.ylabel('Prob density')
plt.title('Distribution plot of age',fontsize=15)
```

```python
sns.set_style("whitegrid")
g = sns.catplot(y= "manufacturer", data = df, kind = "count",
                   palette = "Reds_r", height = 7, aspect = 1.5)
g.ax.set_title("Count of vehicles by its manufacturer",fontdict={'size':15})
g.ax.xaxis.set_label_text("Vehicles",fontdict= {'size':12})
g.ax.yaxis.set_label_text("Manufacturer",fontdict= {'size':12})
plt.show()
```

```python
sns.set_style("whitegrid")
g = sns.catplot(x = "fuel", data = df, kind = "count",
                   palette = "Reds_r", height = 7, aspect = 1.5)
g.ax.set_title("Count of vehicles by its fuel type",fontdict={'size':15})
g.ax.xaxis.set_label_text("fuel type",fontdict= {'size':12})
g.ax.yaxis.set_label_text("vehicle count",fontdict= {'size':12})
plt.show()
```

```python
sns.set_style("whitegrid")
g = sns.catplot(x = "transmission", data = df, kind = "count",
                   palette = "Reds_r", height = 7, aspect = 1.5)
g.ax.set_title("Count of vehicles by its transmission",fontdict={'size':15})
g.ax.xaxis.set_label_text("Vehicles count",fontdict= {'size':12})
g.ax.yaxis.set_label_text("transmission",fontdict= {'size':12})
plt.show()
```

```python
sns.set_style("whitegrid")
g = sns.catplot(x = "title_status", data = df, kind = "count",
                   palette = "Reds_r", height = 7, aspect = 1.5)
g.ax.set_title("Count of vehicles by its title_status",fontdict={'size':15})
g.ax.xaxis.set_label_text("Vehicles count",fontdict= {'size':12})
g.ax.yaxis.set_label_text("title_status",fontdict= {'size':12})
plt.show()
```

```python
sns.set_style("whitegrid")
g = sns.catplot(y= "state_name", data = df, kind = "count",
                   palette = "Reds_r", height = 7, aspect = 1.5)
g.ax.set_title("Count of vehicles by its states",fontdict={'size':15})
g.ax.xaxis.set_label_text("Vehicles count",fontdict= {'size':12})
g.ax.yaxis.set_label_text("state_name",fontdict= {'size':12})
plt.show()
```

```python
sns.lmplot(x = 'age', y = 'price', data = df, palette = 'muted', height = 8)
```

```python
fig, ax = plt.subplots(figsize=(8,5))
colors = ["#1E90FF", "#6495ED", "#B0C4DE"]
sns.barplot(x="fuel", y="price",hue="transmission", palette=colors, data=df)
ax.set_title("price of vehicles by fuel and transmission",fontdict= {'size':12})
ax.xaxis.set_label_text("Fuel type",fontdict= {'size':12})
ax.yaxis.set_label_text("Price",fontdict= {'size':12})
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
colors = ["#1E90FF", "#6495ED", "#B0C4DE"]
sns.barplot(x="title_status", y="price",hue="transmission", palette=colors, data=df)
ax.set_title("price of vehicles by title_status and transmission",fontdict= {'size':12})
ax.xaxis.set_label_text("title_status",fontdict= {'size':12})
ax.yaxis.set_label_text("price",fontdict= {'size':12})
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,15))
colors = ["#1E90FF", "#6495ED", "#B0C4DE"]
sns.barplot(x = "price", y = "state_name",hue = "transmission", palette = colors, data = df)
ax.set_title("price of vehicles by state_name and transmission",fontdict= {'size':12})
ax.xaxis.set_label_text("price",fontdict= {'size':12})
ax.yaxis.set_label_text("state_name",fontdict= {'size':12})
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,10))
colors = ["#6495ED"]
sns.barplot(y = "state_name", x = "price", palette = colors, data = df)
ax.set_title("price of vehicles by state_name",fontdict = {'size':12})
ax.xaxis.set_label_text("price",fontdict = {'size':12})
ax.yaxis.set_label_text("state_name",fontdict = {'size':12})
plt.show()
```

```python
plt.clf()
s = df.pivot_table("price", "manufacturer", aggfunc="mean").plot(kind='bar', figsize = (17,5), rot = 80)
s.axhline(df.price.mean(), color = "red" , linestyle = "-.", label = "MEAN", marker = "D")
s.arrow(20,11000,10,18000, color = 'magenta', head_width = 5, label = 'mean')
s.text(28, 30000, "Mean", horizontalalignment = 'left', size = 'large', color = 'black')
df.pivot_table("price","fuel", aggfunc="mean").plot(kind='bar', figsize=(9,5))
df.pivot_table("price","title_status", aggfunc="mean").plot(kind='bar', figsize=(5,5))
df.pivot_table("price","transmission", aggfunc="mean").plot(kind='bar', figsize=(5,5))
df.pivot_table("price","state_name", aggfunc="mean").plot(kind='bar', figsize=(40,5))
```
